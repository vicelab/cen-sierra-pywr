```python
import os
import json

import glob
import pandas as pd
from tqdm import tqdm
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_path = os.environ['SIERRA_DATA_PATH']
# data_path = r'C:\Users\david\Box\CERC-WET\Task7_San_Joaquin_Model\pywr_models\data'
results_path = os.path.join(data_path, '../results')
whiplash_dir = os.path.join(results_path, 'Whiplash')
basins = [b for b in os.listdir(whiplash_dir)]
basins
```

```python
def load_variable(basin_name, variable):
    path = os.path.join(whiplash_dir, basin_name, 'sequences')
    
    dfs = []
    for sequence in tqdm(os.listdir(path)):
        filepath = os.path.join(path, sequence, variable + '.csv')
        df = pd.read_csv(filepath, index_col=0, header=[0,1], parse_dates=True)
        df = df.get_level
        df = df.resample('M').agg('sum')
        df['sequence'] = sequence
        df = df.reset_index().set_index(['sequence', 'Date'])
        dfs.append(df)
    df = pd.concat(dfs, axis=0)
    return df
```

```python
basin = 'merced'
df_all = load_variable(basin, 'Hydropower_Energy_MWh')
```

```python
df = df_all.sum(axis=1)
df
```

```python
dfm = df.reset_index()
dfm['wy'] = dfm['Date'].map(lambda d: d.year if d.month < 10 else d.year + 1)
dfm.set_index('sequence', 'wy')
dfm.head()
```

```python
df2 = df.sum(axis=1).to_frame()
df2['dry'] = df.index.get_level_values(0).map(lambda x: int(x.split('_')[2].split('W')[0][1:]))
df2['wet'] = df.index.get_level_values(0).map(lambda x: int(x.split('_')[2].split('W')[1]))
df2['n'] = df.index.get_level_values(0).map(lambda x: int(x.split('_')[3][1:]))
df2.reset_index(inplace=True)
del df2['sequence']
df2.set_index(['dry', 'wet', 'n', 'Date'], inplace=True)
df2.columns = ['MWh']
df2.head()
```

```python
df3 = df2.reset_index()
wet1 = df3[df3['wet']==1]
```

```python
dry_years = list(set(df3['dry']))
dry_years
```

```python
D = len(dry_years)
fig, axes = plt.subplots(D, 1, figsize=(15,10))
for i, d in enumerate(dry_years):
    print(d)
    ax = axes[i]
    data = wet1[wet1['dry']==d]
    sns.lineplot(data=data, x='Date', y='MWh', ax=ax)
fig.tight_layout()
```
